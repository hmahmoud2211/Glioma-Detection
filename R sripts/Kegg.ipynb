```python
import gseapy as gp

# Perform enrichment analysis for up genes
enr_up = gp.enrichr(
    gene_list=up_genes,  # List of upregulated genes
    gene_sets='KEGG_2021_Human',  # Choose a library (KEGG, GO, L1000, etc.)
    organism='Human',
    outdir='enrichr_up_results',  # Output folder
)

# Perform enrichment analysis for down genes
enr_down = gp.enrichr(
    gene_list=down_genes,  # List of downregulated genes
    gene_sets='KEGG_2021_Human',  # Choose a library
    organism='Human',
    outdir='enrichr_down_results',  # Output folder
)

# Display top results
print("Top pathways for upregulated genes:")
print(enr_up.results.head())

print("\nTop pathways for downregulated genes:")
print(enr_down.results.head())
```

```python
# Plot the enrichment results for upregulated genes
gp.barplot(enr_up.results, title='Top Pathways for Up Genes', ofname='up_pathways.png')

# Plot for downregulated genes
gp.barplot(enr_down.results, title='Top Pathways for Down Genes', ofname='down_pathways.png')
```
